### Importing Libraries

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading the Dataset

```python
netflix_df = pd.read_csv('./netflix_data.csv')
netflix_df.head()
```

### Filtering the Dataset

```python
# Filtering the data to include movies only
netflix_subset = netflix_df[netflix_df['type']=='Movie']
netflix_subset['type'].value_counts()
```

```python
# Filtering the columns to the ones required for our analysis
netflix_movies = netflix_subset.loc[:, ['title', 'country', 'genre', 'release_year', 'duration']]
netflix_movies.head()
```

```python
# Selecting movies with a runtime of less than 60 minutes
short_movies = netflix_movies[netflix_movies['duration']<60]
short_movies.head()
```

### Plotting Results

```python
# Assigning colors to different genres
colors = []

for label, row in netflix_movies.iterrows() :
    if row["genre"] == "Children" :
        colors.append("yellow")
    elif row["genre"] == "Documentaries" :
        colors.append("black")
    elif row["genre"] == "Stand-Up":
        colors.append("Red")
    else:
        colors.append("green")

print(colors[0:10])
```

```python
# Set the figure style and initalize a new figure
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies.release_year, netflix_movies.duration, c=colors)

# Create a title and axis labels
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")

# Show the plot
plt.show()
```

### Analysis
From the results, it can be observed that netflix movies are not get shorter.
